```python
import torch
x=torch.tensor(3.0,requires_grad=True)
y=x**2
y.backward()
x.grad
```

```python
#Neural Network
import torch
x=torch.tensor(6.7)
y=torch.tensor(1.0)
w=torch.tensor(0.33,requires_grad=True)
bias=torch.tensor(5.0,requires_grad=True)
z=bias+w*x
y_pred=torch.sigmoid(z)
loss=torch.nn.functional.binary_cross_entropy(y_pred,y)
loss.backward()
print(w.grad)
print(bias.grad)
```
