 **SCALAR OPERATION**

```python
import torch
x=torch.rand(2,3)
print(x)
#ADDITION
x+=2
print(x)
#subtraction
x-=2
print(x)
#Multiplication
x*=2
print(x)
#division
x/=2
print(x)
#int division
x=(x*100)//3
print(x)
#mod
x=((x*100)//3)%2
print(x)
#power
x=x**2
print(x)
```

**ELEMENT WISE OPERATION**

```python
import torch
a=torch.rand(3,5)
b=torch.rand(3,5)
#addition
c=a+b
print(c)
#Subtraction
d=a-b
print(d)
#Multiplication
e=a*b
print(e)
#division
f=a/b
print(f)
#power
g=a**b
print(g)
#mod
h=(a*100)//b
print(h)
```
